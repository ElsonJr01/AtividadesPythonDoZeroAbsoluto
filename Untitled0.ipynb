```python
# Importar as bibliotecas necess√°rias
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Dados de treinamento (temperaturas)
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46.4, 59, 71.6, 100], dtype=float)

# Criar o modelo de rede neural
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])

# Compilar o modelo
model.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)

# Treinar o modelo
print("Treinando o modelo...")
hist = model.fit(celsius, fahrenheit, epochs=500, verbose=False)
print("Treinamento conclu√≠do!")

# Plotar gr√°fico do hist√≥rico de treinamento
plt.figure(figsize=(10,5))
plt.plot(hist.history['loss'])
plt.title("Hist√≥rico de Treinamento (Loss por Epoch)")
plt.xlabel("√âpocas")
plt.ylabel("Perda (Loss)")
plt.grid(True)
plt.show()

# Testar com novos dados (n√£o presentes no treinamento)
novas_temperaturas = np.array([-20, -5, 5, 12, 18, 25, 30, 35, 40, 45], dtype=float)
valores_previstos = model.predict(novas_temperaturas)

# Calcular os valores reais para compara√ß√£o
valores_reais = novas_temperaturas * 1.8 + 32

# Criar a tabela comparativa
tabela = pd.DataFrame({
    "Celsius": novas_temperaturas,
    "Previsto (F)": np.round(valores_previstos.flatten(), 2),
    "Real (F)": np.round(valores_reais, 2)
})

print("\nTabela Comparativa - Resultados do Modelo")
print(tabela)
```
